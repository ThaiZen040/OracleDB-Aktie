```python
# Importieren der notwendigen Bibliotheken
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

```

```python
plt.style.use('seaborn-v0_8') # Setzen des Plot-Stils
data = yf.download("ORCL", start="2020-01-01", end="2025-01-01", auto_adjust=0, actions=0)# Herunterladen der Aktienkursdaten von Oracle (ORCL) von Yahoo Finance
data.head() # Anzeigen der ersten fünf Zeilen des DataFrames
```

```python
data = data[['Close']].dropna() # Auswahl der Spalte Close und Entfernen von NaN-Werten
data.rename(columns={'Close': 'Kurs'}, inplace=True) # Umbenennen der Spalte Close in Kurs

print("Datenumfang:", len(data)) # Ausgabe des Datenumfangs
data.describe() # Ausgabe der statistischen Beschreibung der Daten
```

Für die Analyse wird der bereinigte Schlusskurs („Adjusted Close“) verwendet, da dieser – im Gegensatz zum unbereinigten Schlusskurs („Close“) – Dividenden und Splits berücksichtigt. Der Adjusted Close entspricht dem Chart, wie er auf Yahoo Finance dargestellt wird, und bildet die reale Wertentwicklung genauer ab (Yahoo Finance, 2025).

```python
plt.figure(figsize=(12, 5))  # neue Figure mit Größe
plt.plot(data['Kurs'], label='Oracle (Adj Close)') # Plotten der Kursdaten
plt.title('Oracle Corporation – Kursverlauf 2020–2025') # Hinzufügen eines Titels zum Plot
plt.xlabel('Datum') # Hinzufügen eines Labels zur x-Achse
plt.ylabel('Kurs in USD') # Hinzufügen eines Labels zur y-Achse
plt.legend()  # Hinzufügen einer Legende zum Plot
plt.show() # Anzeigen des Plots
```
